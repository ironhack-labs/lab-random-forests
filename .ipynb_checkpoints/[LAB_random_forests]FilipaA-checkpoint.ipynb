- Apply the Random Forests algorithm but this time only by upscaling the data using `SMOTE`.
- Note that since `SMOTE` works on numerical data only, we will first encode the categorical variables in this case.


```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
%matplotlib inline
pd.set_option('display.max_columns', None)

from sklearn.model_selection import train_test_split
```

```python
numerical=pd.read_csv(r"C:\Users\Pedro Gomes\Desktop\Filipa_Ironhack\Labs\ML\lab-random-forests\files_for_lab\numerical.csv")
categorical = pd.read_csv(r"C:\Users\Pedro Gomes\Desktop\Filipa_Ironhack\Labs\ML\lab-random-forests\files_for_lab\categorical.csv")
target = pd.read_csv(r"C:\Users\Pedro Gomes\Desktop\Filipa_Ironhack\Labs\ML\lab-random-forests\files_for_lab\target.csv")
```

```python
categorical
```

```python
# turn this feature into an integer 
categorical['CLUSTER'] = pd.to_numeric(categorical['CLUSTER'])
categorical.dtypes
```

```python
# since there are a lot of nuemrical variables in the categorical df, and 
# i dont want to encode them all I have to concate categorical and numerical separate them again enconde and then 
# concat again
data_1 = pd.concat([numerical, categorical], axis=1)
data_1
```

```python
# New dataframe with just categorical values
categorical2 = data_1.select_dtypes(object)
categorical2.dtypes
```

```python
# Encode the whole dataframe
categorical_encoded = pd.get_dummies(categorical2, columns=categorical2.columns)
categorical_encoded
```

```python
# concat numerical dataframe and the categorical_encoded
data = pd.concat([numerical, categorical_encoded], axis = 1)
data
```

```python
# there are no null values 
data.isna().sum()/len(data)
```

```python
# take the target D of the target df we only need target B
target =target.drop(['TARGET_D'], axis=1)
```

```python
# Check for data imbalace 
target.value_counts()
```

```python
# First lets splitt the data, because we only want to oversample in the train data
# since we already have the data set separated from the target valariable we will use it directly to do the train test split 

# Doing the train-test-split

X_train, X_test, y_train, y_test = train_test_split(data, target, test_size = 0.2, random_state=0)
# to do the SMOTE we have to concat the the 2 train data sets
train = pd.concat([X_train, y_train],axis=1)
```

```python
# SMOTE

category_0 = train[train['TARGET_B']==0]
category_1 = train[train['TARGET_B']==1]

from imblearn.over_sampling import SMOTE 

smote = SMOTE()
X = train.drop('TARGET_B', axis=1)
y = train['TARGET_B']

X_sm, y_sm = smote.fit_resample(X,y)
y_sm.value_counts()
```

```python
# Concat the data to train the model with the train_oversampled data
train_SMOTE = pd.concat([X_sm, y_sm], axis = 1)

train_SMOTE
```

```python
# Define the X and the y of the oversampled train data

X_train_SMOTE = train_SMOTE.drop(['TARGET_B'], axis = 1)
y_train_SMOTE = train_SMOTE['TARGET_B']

from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(n_estimators = 50, max_depth = 7, random_state = 42)
clf.fit(X_train_SMOTE, y_train_SMOTE)

print(clf.score(X_train_SMOTE, y_train_SMOTE))
print(clf.score(X_test, y_test))
```

```python
from sklearn.model_selection import GridSearchCV
# Lets Tune the hyperparameters

param_grid = {'n_estimators': [50, 100, 200],
              'max_depth': [2, 4, 7],
              'min_samples_split': [2,4],
              'min_samples_leaf': [1,2]}

# secondly define the model you wnat 
clf = RandomForestClassifier(random_state = 42)
grid_search = GridSearchCV(clf, param_grid, cv=5, return_train_score=True, n_jobs = -1 )
# cv is the crosso validation 
# return train score, should be true to cheeck for overfitting

grid_search.fit(X_train_SMOTE, y_train_SMOTE)
grid_search.best_params_
```

```python
# Build a model with the best hyperparameters
clf = RandomForestClassifier(n_estimators = 50, max_depth = 7, random_state = 42)
clf.fit(X_train_SMOTE, y_train_SMOTE)

print(clf.score(X_train_SMOTE, y_train_SMOTE))
print(clf.score(X_test, y_test))
```
