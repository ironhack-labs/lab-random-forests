```python
categorical = pd.read_csv('/Users/leozinho.air/Desktop/Ironhack/class_30/lab-random-forests/files_for_lab/categorical.csv')
numerical = pd.read_csv('/Users/leozinho.air/Desktop/Ironhack/class_30/lab-random-forests/files_for_lab/numerical.csv')
target = pd.read_csv('/Users/leozinho.air/Desktop/Ironhack/class_30/lab-random-forests/files_for_lab/target.csv')
```

```python
numerical2 = categorical.select_dtypes(include=np.number)
numerical2
```

```python
categorical = categorical.select_dtypes(include= object)
categorical
```

```python
dummies = pd.get_dummies(categorical, drop_first = True)
dummies
```

```python
categorical_X = pd.concat([dummies,numerical2], axis=1)
categorical_X
```

```python
data = pd.concat([numerical, categorical_X, target.drop("TARGET_D", axis = 1) ], axis = 1).reset_index(drop=True)
data
```

```python
from sklearn.model_selection import train_test_split

# Defining X and y variables
X = data.drop("TARGET_B",axis = 1)
y = data["TARGET_B"]

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42, test_size = 0.2)

# Train dataframe
df_train = pd.concat([X_train, y_train], axis = 1)
df_train = df_train.reset_index(drop = True)


# Test dataframe
df_test = pd.concat([X_test, y_test], axis = 1)
df_test = df_test.reset_index(drop = True)
```

```python

# Computing the frequency of the unique values of the target variable before oversampling or undersampling the train dataset
df_train["TARGET_B"].value_counts()
```

```python
# Upscaling with SMOTE
from imblearn.over_sampling import SMOTE

smote = SMOTE()
X_sm, y_sm = smote.fit_resample(df_train.drop("TARGET_B", axis = 1), df_train["TARGET_B"])

# Creating a balanced train dataframe
df_train = pd.concat([X_sm, y_sm], axis = 1)
df_train
```

```python
df_train["TARGET_B"].value_counts()
```

```python
from sklearn.ensemble import RandomForestClassifier

# Building a Random Forest model

rf_model = RandomForestClassifier(n_estimators = 100, max_depth = 7, random_state = 42)
rf_model.fit(df_train.drop("TARGET_B", axis = 1), df_train["TARGET_B"])

print("R2 (Train data):", rf_model.score(df_train.drop("TARGET_B", axis = 1), df_train["TARGET_B"]))
print("R2 (Test data):", rf_model.score(df_test.drop("TARGET_B", axis = 1), df_test["TARGET_B"]))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report

# Make predictions on the test data
y_pred_test = rf_model.predict(df_test.drop("TARGET_B", axis=1))

# Make predictions on the training data for comparison
y_pred_train = rf_model.predict(df_train.drop("TARGET_B", axis=1))

# Evaluate the model
# Test data metrics
accuracy_test = accuracy_score(df_test["TARGET_B"], y_pred_test)
precision_test = precision_score(df_test["TARGET_B"], y_pred_test, average='binary')
recall_test = recall_score(df_test["TARGET_B"], y_pred_test, average='binary')
f1_test = f1_score(df_test["TARGET_B"], y_pred_test, average='binary')

# Training data metrics for comparison
accuracy_train = accuracy_score(df_train["TARGET_B"], y_pred_train)
precision_train = precision_score(df_train["TARGET_B"], y_pred_train, average='binary')
recall_train = recall_score(df_train["TARGET_B"], y_pred_train, average='binary')
f1_train = f1_score(df_train["TARGET_B"], y_pred_train, average='binary')

# Print evaluation metrics
print("Evaluation Metrics for Training Data:")
print(f"Accuracy (Train): {accuracy_train}")
print(f"Precision (Train): {precision_train}")
print(f"Recall (Train): {recall_train}")
print(f"F1 Score (Train): {f1_train}")

print("\nEvaluation Metrics for Test Data:")
print(f"Accuracy (Test): {accuracy_test}")
print(f"Precision (Test): {precision_test}")
print(f"Recall (Test): {recall_test}")
print(f"F1 Score (Test): {f1_test}")

# Confusion Matrix for Test Data
conf_matrix_test = confusion_matrix(df_test["TARGET_B"], y_pred_test)
print("\nConfusion Matrix for Test Data:")
print(conf_matrix_test)

# Detailed classification report for Test Data
```
